```python
#. Checking data types
marketing_spend.info()
```

```python
# a. Converting to Categorical to save memory and improve speed
obj_cols = ['platform']
for col in obj_cols:
    marketing_spend[col] = marketing_spend[col].astype('category')
```

```python
# b. Downcasting float Columns to memory-efficient types
# Specific smaller-range integers
flo_cols = ['cpc', 'cpa', 'roas']
marketing_spend[flo_cols] = marketing_spend[flo_cols].astype('float32')
```

```python
# c. Ensuring the financial metrics columns are float64 for aggregation precision
float64_cols = ['spend', 'revenue_attributed']
for col in float64_cols:
    marketing_spend[col] = marketing_spend[col].astype('float64')
```

```python
# d. Ensuring the other metrics columns are int64
int64_cols = ['clicks', 'conversions', 'impressions']
for col in int64_cols:
    marketing_spend[col] = marketing_spend[col].astype('int64')

print(marketing_spend.dtypes)
```

```python
import pandas as pd
from datetime import datetime

# Assuming your DataFrame has a column named 'month' with date-like strings
# First, convert the object column to datetime
marketing_spend['month'] = pd.to_datetime(marketing_spend['month'])

# Extract year and month into separate columns
marketing_spend['year'] = marketing_spend['month'].dt.year
marketing_spend['month_num'] = marketing_spend['month'].dt.month

# Alternative: Extract year-month as a string (YYYY-MM format)
marketing_spend['year_month'] = marketing_spend['month'].dt.to_period('M')

# If your original column has different formats, you might need to specify the format:
# df['month'] = pd.to_datetime(df['month'], format='%Y-%m-%d')  # adjust format as needed

# Example with sample data:
# df = pd.DataFrame({'month': ['2023-01-15', '2023-02-20', '2024-03-10']})
```

```python
import pandas as pd
import sqlite3

# 1. Establishing a connection
# This creates the file 'marketing_spend.db' in the working directory
with sqlite3.connect('marketing_spend.db') as conn:

    # Convert Period columns to strings before writing to SQL
    # This handles the Period data type that SQLite doesn't support
    marketing_spend_copy = marketing_spend.copy()
    
    # More robust conversion of Period columns to string format
    for col in marketing_spend_copy.columns:
        # Check if column contains Period data type (more comprehensive check)
        if hasattr(marketing_spend_copy[col].dtype, 'name') and 'period' in str(marketing_spend_copy[col].dtype).lower():
            marketing_spend_copy[col] = marketing_spend_copy[col].astype(str)
        # Additional check for individual Period objects in the column
        elif marketing_spend_copy[col].apply(lambda x: str(type(x).__name__) == 'Period').any():
            marketing_spend_copy[col] = marketing_spend_copy[col].astype(str)
    
    # 2. Writing the data to a table
    # 'name' is the table name for the db
    # 'if_exists' replaces the table if it already exists
    marketing_spend_copy.to_sql(name='marketing_spend', con=conn, if_exists='replace', index=False)

# 3. To make sure the database is correctly built before downloading it
import os
file_stats = os.stat('marketing_spend.db')
print(f"Database size: {file_stats.st_size / (1024 * 1024):.2f} MB")
```
