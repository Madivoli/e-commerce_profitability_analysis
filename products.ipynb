```python
# 1. Importing the dataset
```

```python
import pandas as pd

products = pd.read_csv(r"G:\My Drive\Data Analysis Projects\E-Commerce\Dataset\products.csv")

print(products.head())
```

```python
# 2. Checking and addressing missing and NaN values
```

```python
# a. Checking for missing values
products.isna().sum()
```

```python
# 3. Checking and converting data types:
```

```python
#. Checking data types
products.info()
```

```python
# b. Converting to Categorical to save memory and improve speed
cat_cols = [ 'product_name', 'category', 'sub_category', 'supplier']
for col in cat_cols:
    products[col] = products[col].astype('category')
```

```python
# c. Downcasting operational financial columns to float32
float64_cols = [
    'unit_cost', 'selling_price', 'shipping_cost_per_unit', 
    'weight_lbs'
]
for col in float64_cols:
    products[col] = products[col].astype('float32')
```

```python
print(products.dtypes)
```

```python
# 4. Checking for duplicates
```

```python
duplicate_rows = products.duplicated()
print(f"Number of duplicate rows: {duplicate_rows.sum()}")

duplicates = products[products.duplicated()]
print(duplicates)
```

```python
# 5. Data transformation (trimming)
```

```python
import re

# Identify object columns
obj_cols = products.select_dtypes(include=['object']).columns

for col in obj_cols:
# Trimming Whitespaces
    products[col] = products[col].str.strip()
```

```python
# 6. Resetting the table's index column
```

```python
products = products.reset_index(drop=True)

print(products.head())
print("Cleaning Process Complete!")
print(f"Total Rows: {len(products)}")
print(f"Null Values Remaining: {products.isnull().sum().sum()}")
```

```python
# CREATING AND SAVING A CLEAN CSV FILE
```

```python
products.to_csv(r"G:\My Drive\Data Analysis Projects\E-Commerce\Cleaned\products.csv", index=False)

import os
file_path = r"G:\My Drive\Data Analysis Projects\E-Commerce\Cleaned\products.csv"
if os.path.exists(file_path):
    print("File successfully created!")
```

```python
CREATING AND SAVING DATABASE
```

```python
import pandas as pd
import sqlite3


# 1. Establishing a connection
# This creates the file 'products.db' in the working directory
with sqlite3.connect('products.db') as conn:

# 2. Writing the data to a table
# 'name' is the table name for the db
# 'if_exists' replaces the table if it already exists
    products.to_sql(name='products', con=conn, if_exists='replace', index=False)

# 3. To make sure the database is correctly built before downloading it
import os
file_stats = os.stat('products.db')
print(f"Database size: {file_stats.st_size / (1024 * 1024):.2f} MB")
```
